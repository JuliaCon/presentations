# React.jl

General purpose signals library by Shashi Gowda, inspired by Elm and functional reactive programming.

```julia
using React, React.Timing, Interact, IJuliaWidgets, RDatasets, Gadfly
```

```julia
r = Input(1)
xs = rand(10)
lift(r -> xs .^ r, r)
```

```julia
push!(r, 20)
```

```julia
diamonds = dataset("ggplot2", "diamonds")
set_default_plot_size(6inch, 4inch)
plot(diamonds, x=:Carat, y=:Price, Scale.y_log10, Geom.hexbin)
```

```julia
xbincount = Input(100)
Slider(xbincount, "X bincount", 100, 10, 200, 10)
```

```julia
ybincount = Input(100)
Slider(ybincount, "Y bincount", 100, 10, 200, 10)
```

```julia
lift(xbincount -> plot(sin, 0, 1 + xbincount), xbincount)
```

# Three Line Theramin

Macbook ambient light sensor + React.jl + AudioIO = Theramin

```julia
using AudioIO

ambient_light = lift(t -> ccall((:readlmu, :lmu), Uint64, ()) / 1e7, fps(20.0))
osc = play(SinOsc(440.0))
lift(l -> osc.renderer.freq = 440.0 * l, ambient_light)
```
